```python
import cv2
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=cv2.imread('./testset/rcj.jpg')
plt.imshow(img)
```

```python
face_cascade=cv2.CascadeClassifier('./opencv/haarcascades/haarcascade_eye.xml')
face=face_cascade.detectMultiScale(img,1.5,5)
face
```

```python
for (x,y,w,h) in face:
    face_det=cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    plt.imshow(face_det)
```
